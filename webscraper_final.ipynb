```python
#below we will build a series of for loops to scrape each page in sequence for the desired data category. 
#only the first instances of these steps will have comments. 
```

```python
loc_list = []

for x in range (1,68):
    url = 'https://www.coldwellbankertamarindo.com/single-family-homes-for-sale?page='
    r = requests.get(url+str(x))
    soup = BeautifulSoup(r.content, 'html.parser') 
    content= soup.find_all('div', {'class': 'f-14 c-lighter fix-line'})
    (loc_list.append(content))
        

    
    
        
        
```

```python
loc_list
```

```python
#the brackets indicate we have a series of lists returned, however we need them in a single list to form our dataframe or else
#  each different location will form a column, not a place in a single row. 
#we convert the list series to a flat list with another for loop. 
```

```python
loc_flat_list = []
 
for sublist in loc_list:
    for item in sublist:
        loc_flat_list.append(item)
```

```python
loc_flat_list
```

```python
#now we can convert our single (flat) list into a dataframe. 
```

```python
loc_df = pd.DataFrame(loc_flat_list)
```
